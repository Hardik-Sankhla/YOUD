```python
from pytube import YouTube
from IPython.display import display, HTML, clear_output
import ipywidgets as widgets
import os

DOWNLOADS_DIR = "/content/Downloads"  # Change this path as needed

# Function to download video
def download_video(link):
    try:
        yt = YouTube(link)
        stream = yt.streams.get_highest_resolution()
        video_dir = os.path.join(DOWNLOADS_DIR, "video")
        os.makedirs(video_dir, exist_ok=True)
        stream.download(output_path=video_dir)
        output_text.value += f"<span style='color: #27ae60;'>Video '{yt.title}' downloaded successfully!</span><br>"
    except Exception as e:
        output_text.value += f"<span style='color: #c0392b;'>An error occurred: {str(e)}</span><br>"

# Function to download audio
def download_audio(link):
    try:
        yt = YouTube(link)
        stream = yt.streams.filter(only_audio=True).first()
        audio_dir = os.path.join(DOWNLOADS_DIR, "audio")
        os.makedirs(audio_dir, exist_ok=True)
        stream.download(output_path=audio_dir)
        output_text.value += f"<span style='color: #27ae60;'>Audio '{yt.title}' downloaded successfully!</span><br>"
    except Exception as e:
        output_text.value += f"<span style='color: #c0392b;'>An error occurred: {str(e)}</span><br>"

# Function to process links for download
def process_links(links, download_option):
    for link in links:
        if download_option == 'Video':
            download_video(link)
        elif download_option == 'Audio':
            download_audio(link)

# Function to add links and trigger download
def add_links(button):
    links_text = links_input.value.strip()
    if links_text:
        links = links_text.split('\n')
        download_option = download_dropdown.value
        process_links(links, download_option)
        clear_output_area(None)
        output_text.value += f"<strong>Downloaded {len(links)} links.</strong><br>"

# Function to clear the output area
def clear_output_area(button):
    output_text.value = ""

# Custom CSS for styling
custom_css = """
<style>
.output-area {
    border: 1px solid #ccc;
    border-radius: 5px;
    padding: 8px;
    font-size: 14px;
    margin-bottom: 10px;
    background-color: #f5f5f5;
    max-height: 300px;
    overflow: auto;
    animation: fadeInAnimation ease 3s;
}

@keyframes fadeInAnimation {
    0% {
        opacity: 0;
    }
    100% {
        opacity: 1;
    }
}

.download-button {
    background: linear-gradient(to bottom, #3498db, #2980b9);
    color: white;
    padding: 10px 20px;
    text-align: center;
    text-decoration: none;
    display: inline-block;
    font-size: 16px;
    margin: 10px 5px;
    cursor: pointer;
    border-radius: 5px;
    transition: all 0.3s ease-in-out;
    border: 2px solid #2980b9;
}

.download-button:hover {
    background: linear-gradient(to bottom, #2980b9, #1f6690);
}

.custom-title {
    font-size: 28px;
    color: #2c3e50;
    margin: 0 0 20px 0;
    text-align: center;
}

.custom-container {
    background: linear-gradient(to right, #e74c3c, #3498db);
    padding: 20px;
    border-radius: 10px;
    box-shadow: 0px 0px 20px 0px rgba(0, 0, 0, 0.1);
}

.custom-dropdown {
    width: 250px;
    margin: 0 auto;
    padding: 10px;
    border-radius: 5px;
    background: linear-gradient(to bottom, #3498db, #2980b9);
    color: white;
}

.custom-textarea {
    border: 2px solid #3498db;
    border-radius: 5px;
    padding: 8px;
    font-size: 14px;
    margin-bottom: 10px;
    color: #3498db;
}
</style>
"""

# Widgets creation
links_input = widgets.Textarea(placeholder='Enter YouTube video links (one per line)', layout={'height': '150px'}, add_class='custom-textarea')
download_dropdown = widgets.Dropdown(options=['Video', 'Audio'], description='Download Option', layout={'width': 'initial'}, style={'description_width': 'initial'}, add_class='custom-dropdown')
add_links_button = widgets.Button(description='Download Links', layout=widgets.Layout(flex='1 1 auto'), add_class='download-button')
clear_output_button = widgets.Button(description='Clear Output', layout=widgets.Layout(flex='1 1 auto'), add_class='download-button')
output_text = widgets.HTML(value='', layout=widgets.Layout(min_height='300px'), placeholder='Output...', description='Output:', disabled=False, add_class='output-area')

# Event handlers
add_links_button.on_click(add_links)
clear_output_button.on_click(clear_output_area)

# HTML layout creation
html_title = widgets.HTML(value="<h2 class='custom-title'>YouTube Downloader</h2>")
button_section = widgets.HBox([add_links_button, clear_output_button])
output_section = widgets.VBox([output_text])
widgets_section = widgets.VBox([links_input, download_dropdown, button_section, output_section], layout=widgets.Layout(align_items='center'))

# Combine widgets and HTML layout
main_container = widgets.VBox([html_title, widgets_section], layout=widgets.Layout(justify_content='center'))
main_container.add_class('custom-container')

# Display UI with custom styling
display(HTML(custom_css))
display(main_container)
```
